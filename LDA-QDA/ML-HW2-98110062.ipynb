Linear Discriminant Analysis

```python
class MyLDA:    
    def fit(self, X, y):        
        n_samples, n_features = X.shape
        self.unique_y_values, unique_y_counts = np.unique(y, return_counts = True)
        self.pi_ks = unique_y_counts / n_samples
        self.mu_ks = []
        self.Sigma = np.zeros((n_features, n_features))        
        for k in self.unique_y_values:
            X_k = X[y == k]
            mu_k = np.mean(X_k, axis=0).reshape(n_features, 1)
            self.mu_ks.append(mu_k)
            for sample in X_k:
                sample = sample.reshape(-1,1)
                difference = sample - mu_k
                self.Sigma += np.dot(difference, difference.T) 
        self.Sigma /= n_samples
        
    def calculate_density(self,sample, mu_k, Sigma):
        difference = sample - mu_k
        Sigma_regularizated = Sigma + 0.5 * np.eye(Sigma.shape[0])
        density = np.exp(-(1/2)* difference.T @ np.linalg.inv(Sigma_regularizated) @ difference)
        return density
            
    def predict(self, X_test):
        y_pred = []
        for sample in X_test:
            sample = sample.reshape(-1, 1)
            p_ks = np.empty(len(self.unique_y_values))
            for j in range(len(self.unique_y_values)):
                p_x_given_y = self.calculate_density(sample, self.mu_ks[j], self.Sigma)
                p_y_given_x = self.pi_ks[j] * p_x_given_y
                p_ks[j] = p_y_given_x
            
            y_pred.append(self.unique_y_values[np.argmax(p_ks)])
        return np.array(y_pred)
```

Quadratic Discriminant Analysis (Optional) (15 bonus points)

```python
class MyQDA:

    def fit(self, X, y):
        n_samples, n_features = X.shape 
        self.unique_y_values, unique_y_counts = np.unique(y, return_counts = True)
        self.pi_ks = unique_y_counts / n_samples        
        self.mu_ks = []
        self.Sigma_ks = []
        for k in self.unique_y_values:
            X_k = X[y == k]
            mu_k = np.mean(X_k, axis=0).reshape(n_features, 1)
            self.mu_ks.append(mu_k)
            Sigma_k = np.zeros((n_features, n_features))
            for sample in X_k:
                sample = sample.reshape(-1,1)
                difference = sample - mu_k
                Sigma_k += np.dot(difference, difference.T)
            self.Sigma_ks.append(Sigma_k/len(X_k))
     
    def calculate_density(self, sample, mu_k, Sigma_k):
        difference = sample - mu_k
        Sigma_regularizated = Sigma_k + 0.75 * np.eye(Sigma_k.shape[0])
        density = np.linalg.det(Sigma_regularizated)**(-1/2) * np.exp(-(1/2) * difference.T @ np.linalg.inv(Sigma_regularizated) @ difference)
        return density
    
    def predict(self, X_test):
        y_pred = []
        for sample in X_test:
            sample = sample.reshape(-1, 1)
            p_ks = np.empty(len(self.unique_y_values))
            for j in range(len(self.unique_y_values)):
                p_x_given_y = self.calculate_density(sample, self.mu_ks[j], self.Sigma_ks[j])
                p_y_given_x = self.pi_ks[j] * p_x_given_y
                p_ks[j] = p_y_given_x
            
            y_pred.append(self.unique_y_values[np.argmax(p_ks)])
        
        return np.array(y_pred)
```

```python
my_qda = MyQDA()
my_qda.fit(x_train, y_train)
```
